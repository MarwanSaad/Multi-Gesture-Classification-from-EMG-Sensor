```python
def evaluate_model(x : np.array, y : np.array ) -> Tuple[np.array, np.array]:
    k_fold = StratifiedKFold(10)

    predicted_targets = np.empty([0], dtype=int)
    actual_targets = np.empty([0], dtype=int)
    accuracy_list = list()

    for train_ix, test_ix in k_fold.split(x,y):
        train_x, train_y, test_x, test_y = x[train_ix], y[train_ix], x[test_ix], y[test_ix]

        # Fit the classifier
        classifier = model3.fit(train_x, train_y)

        # Predict the labels of the test set samples
        predicted_labels = classifier.predict(test_x)
        accuracy = accuracy_score(test_y, predicted_labels)

        predicted_targets = np.append(predicted_targets, predicted_labels)
        actual_targets = np.append(actual_targets, test_y)
        accuracy_list.append(accuracy)


    return predicted_targets, actual_targets, accuracy_list
```

```python
def plot_confusion_matrix(actual_classes : np.array, predicted_classes : np.array):

    matrix = confusion_matrix(actual_classes, predicted_classes)
    
    plt.figure(figsize=(12.8,6))
    sns.heatmap(matrix, annot=True, cmap="Oranges", fmt="g")
    plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('Confusion Matrix')

    plt.show()
```

```python
predicted_target, actual_target, accuracy_list = evaluate_model(x.to_numpy(), y.to_numpy())
print("Mean accuracy:", np.mean(accuracy_list))
#actual_classes, predicted_classes, _ = cross_val_predict(model, kfold, x.to_numpy(), y.to_numpy())
```
